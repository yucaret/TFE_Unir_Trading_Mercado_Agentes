```python
import openai
import csv
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report

# ========================================
# Configuración principal
# ========================================

openai.api_key = ""  # Reemplaza con tu clave real de OpenAI

# Modelos a comparar (solo ChatCompletion)
MODELOS = [
    "gpt-3.5-turbo",  # GPT-3.5
    "gpt-3.5-turbo-16k",
    "gpt-4",           # GPT-4
    "gpt-4-turbo"
]

# Plantilla de prompt:
# El modelo debe responder estrictamente con "positivo", "negativo" o "neutral".
PROMPT_TEMPLATE = """Clasifica el siguiente texto relacionado con el mercado financiero como 'positivo', 'negativo' o 'neutral':
Texto: "{texto}"

Responde únicamente con una de estas palabras: positivo, negativo o neutral.
"""
```

```python
# ========================================
# Funciones
# ========================================

def clasificar_texto_chat(texto, modelo):
    """
    Llama a la API ChatCompletion con 'modelo' (gpt-3.5-turbo o gpt-4).
    Retorna la etiqueta predicha: 'positivo', 'negativo', o 'neutral'.
    """
    prompt = PROMPT_TEMPLATE.format(texto=texto)

    response = openai.chat.completions.create(
        model=modelo,
        messages=[{"role": "user", "content": prompt}],
        max_tokens=5,
        temperature=0
    )

    etiqueta = response.choices[0].message.content.strip().lower()

    # Si no cumple exactamente con la etiqueta, forzamos 'neutral' o
    # podrías usar otra lógica:
    if etiqueta not in ["positivo", "negativo", "neutral"]:
        etiqueta = "neutral"

    return etiqueta
```

```python
def cargar_dataset(ruta_csv):
    """
    Lee el CSV y devuelve una lista de tuplas (texto, etiqueta_real).
    Suponemos que la columna se llama 'Texto_original' y 'Sentimiento'.
    """
    dataset = []
    with open(ruta_csv, mode="r", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        for row in reader:
            texto = row["Texto_original"]
            etiqueta_real = row["Sentimiento"]
            dataset.append((texto, etiqueta_real))
    return dataset
```

```python
def evaluar_modelo(modelo, dataset):
    """
    Para cada texto en el dataset, llama a 'clasificar_texto_chat' y
    devuelve la lista de predicciones.
    """
    predicciones = []
    for (texto, etiqueta_real) in dataset:
        etiqueta_pred = clasificar_texto_chat(texto, modelo)
        predicciones.append(etiqueta_pred)
        #print("texto :", texto, " real: ", etiqueta_real, " predicha: ", etiqueta_pred)
    return predicciones
```

```python

def main():
    # 1. Cargar dataset
    ruta_dataset = "dataset.csv"
    data = cargar_dataset(ruta_dataset)
    textos = [x[0] for x in data]
    labels_reales = [x[1] for x in data]

    # Lista para almacenar resultados por modelo
    resultados_modelos = []

    # 2. Evaluar cada modelo (gpt-3.5-turbo, gpt-4)
    for modelo in MODELOS:
        print(f"\n=== Evaluando modelo: {modelo} ===")
        predicciones = evaluar_modelo(modelo, data)

        # 3. Calcular reporte
        # output_dict=True para extraer métricas en formato dict
        report_dict = classification_report(
            labels_reales, predicciones,
            digits=3, output_dict=True
        )

        # Imprimimos reporte textual para ver en consola
        reporte_textual = classification_report(labels_reales, predicciones, digits=3)
        print(reporte_textual)

        # Extraemos Accuracy, Macro Precision, Macro Recall, Macro F1
        accuracy = report_dict['accuracy']
        macro_precision = report_dict['macro avg']['precision']
        macro_recall = report_dict['macro avg']['recall']
        macro_f1 = report_dict['macro avg']['f1-score']

        resultados_modelos.append({
            "Modelo": modelo,
            "Accuracy": accuracy,
            "Precision (Macro)": macro_precision,
            "Recall (Macro)": macro_recall,
            "F1-score (Macro)": macro_f1
        })

    # 4. Mostramos en tabla y graficamos
    df_resultados = pd.DataFrame(resultados_modelos)
    print("\nResumen de métricas por modelo:")
    print(df_resultados)

    # Convertimos en un DF con índice=Modelo y columnas las métricas
    df_plot = df_resultados.set_index("Modelo")[["Accuracy", "Precision (Macro)", "Recall (Macro)", "F1-score (Macro)"]]

    # Hacemos un gráfico de barras
    plt.figure(figsize=(10,6))
    df_plot.plot(kind='bar', rot=0)
    plt.title("Comparación de métricas entre Modelos Chat GPT")
    plt.xlabel("Modelos")
    plt.ylabel("Valor (0-1)")
    plt.ylim(0, 1.0)
    plt.legend(loc="lower right")
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
