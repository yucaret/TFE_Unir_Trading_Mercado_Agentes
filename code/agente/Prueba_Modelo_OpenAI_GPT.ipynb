#Configuramos el tokken de permiso para acceso al modelo

```python
import openai
import os

apikey_openai = ''
os.environ['OPENAI_API_KEY'] = apikey_openai

openai.api_key = apikey_openai
```

#Definimos una función para cargar los modelo OpenAI

```python
def test_openai_model(model_name, prompt, max_tokens=500):
    response = openai.chat.completions.create(
        model=model_name,
        messages=[
            {"role": "system",
             "content": "Eres un asistente experto en análisis financiero."},
            {"role": "user",
             "content": prompt},
        ],
        max_tokens=max_tokens,
        temperature=0.7,
    )
    return response.choices[0].message.content.strip()
```

#Prueba de los Modelos

```python
import time

prompt = "Explícame cómo los modelos de lenguaje largo pueden mejorar los análisis financieros."

# Probar GPT-3.5
print("GPT-3.5:")
start_time = time.time()
print(test_openai_model("gpt-3.5-turbo", prompt))
end_time = time.time()
print(f"Tiempo de generación: {end_time - start_time:.2f} segundos")
print("")
print("")

# Probar GPT-3.5 Turbo-16k
print("GPT-3.5 Turbo-16k:")
start_time = time.time()
print(test_openai_model("gpt-3.5-turbo-16k", prompt))
end_time = time.time()
print(f"Tiempo de generación: {end_time - start_time:.2f} segundos")
print("")
print("")

# Probar GPT-4
print("GPT-4:")
start_time = time.time()
print(test_openai_model("gpt-4", prompt))
end_time = time.time()
print(f"Tiempo de generación: {end_time - start_time:.2f} segundos")
print("")
print("")

# Probar GPT-4 Turbo
print("GPT-4 Turbo:")
start_time = time.time()
print(test_openai_model("gpt-4-turbo", prompt))
end_time = time.time()
print(f"Tiempo de generación: {end_time - start_time:.2f} segundos")
```
