#Cargamos el modelo gpt-neox-20b

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Nombre del modelo en Hugging Face
model_name = "EleutherAI/gpt-neox-20b"

# Descargar el modelo y el tokenizador
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", torch_dtype="auto")

print("Modelo GPT-NeoX cargado exitosamente")
```

#Prueba del Modelo

```python
# Verifica que el modelo está en la GPU
print(model.device)
```

```python
import time

# Prompt de ejemplo
prompt = "Explícame cómo los modelos de lenguaje largo pueden mejorar los análisis financieros."

# Tokenizar el prompt
inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

start_time = time.time()

# Generar texto
outputs = model.generate(
    inputs.input_ids,
    max_length=50,  # Limita la longitud máxima de la salida
    temperature=0.7,  # Controla la creatividad
    top_k=50,         # Usa las 50 mejores opciones
    top_p=0.9         # Probabilidad acumulativa para filtrar opciones
)

end_time = time.time()

# Decodificar y mostrar el resultado
response = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(response)

print(f"Tiempo de generación: {end_time - start_time:.2f} segundos")
```
