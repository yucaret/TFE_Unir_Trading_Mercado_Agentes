#Cargamos el modelo Llama-2-7b-chat-hf

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Nombre del modelo en Hugging Face
model_name = "meta-llama/Llama-2-7b-chat-hf"

# Descargar el modelo y el tokenizador
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto")

print("Modelo y tokenizador cargados con éxito")
```

#Prueba del Modelo

```python
# Verifica que el modelo está en la GPU
print(model.device)
```

```python
import time

# Prompt de ejemplo
prompt = "Explícame cómo los modelos de lenguaje largo pueden mejorar los análisis financieros."

# Tokenizar el prompt
inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

start_time = time.time()

# Generar texto
outputs = model.generate(
    inputs.input_ids,
    max_length=50,
    ##max_length=100,
    ##max_length=200,
    temperature=0.7,
    top_k=50,  # Considera solo las 50 mejores opciones
    top_p=0.9  # Filtra opciones con probabilidad acumulada menor al 90%
)

end_time = time.time()

# Decodificar y mostrar el resultado
response = tokenizer.decode(outputs[0], skip_special_tokens=True)

print(response)

print(f"Tiempo de generación: {end_time - start_time:.2f} segundos")
```
